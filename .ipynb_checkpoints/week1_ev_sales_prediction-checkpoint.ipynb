```python
# Week 1 - Electric Vehicle Sales Prediction
# Author: Shreya V
# Internship: EduNet AICTE Internship (Electric Vehicle Theme)

# -----------------------------
# Step 1: Import Libraries
# -----------------------------
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import Image, display 

plt.style.use("seaborn-v0_8")
sns.set_palette("Set2")

# -----------------------------
# Step 2: Load Dataset
# -----------------------------
df = pd.read_csv("IEA-EV-dataEV salesHistoricalCars.csv")

# Display first few rows
print("ðŸ”¹ Preview of dataset:")
display(df.head())

# -----------------------------
# Step 3: Understand the Dataset
# -----------------------------
print("\nðŸ”¹ Dataset Info:")
df.info()

print("\nðŸ”¹ Statistical Summary:")
display(df.describe())

print("\nðŸ”¹ Missing Values:")
display(df.isnull().sum())

# -----------------------------
# Step 4: Check Column Names
# -----------------------------
print("\nðŸ”¹ Column Names:")
print(df.columns.tolist())
```

```python
# Step 5: Data Cleaning
# Filter only EV sales rows
df_sales = df[df["parameter"] == "EV sales"]

# Remove unnecessary columns
df_sales = df_sales[["region", "year", "value"]]

# Convert year and value to numeric types
df_sales["year"] = pd.to_numeric(df_sales["year"], errors="coerce")
df_sales["value"] = pd.to_numeric(df_sales["value"], errors="coerce")

# Drop any missing values
df_sales = df_sales.dropna()

print("âœ… Cleaned dataset shape:", df_sales.shape)
display(df_sales.head())
```

```python
# Step 6: Exploratory Data Analysis (EDA)

# EV Sales over the years (global)
plt.figure(figsize=(10,5))
sns.lineplot(data=df_sales, x="year", y="value")
plt.title("Global Electric Vehicle Sales Over the Years")
plt.xlabel("Year")
plt.ylabel("EV Sales (in millions)")
plt.show()

# EV Sales by Region (Top 5 regions)
top_regions = df_sales.groupby("region")["value"].sum().sort_values(ascending=False).head(5)
plt.figure(figsize=(8,4))
sns.barplot(x=top_regions.index, y=top_regions.values)
plt.title("Top 5 Regions by Total EV Sales")
plt.ylabel("Sales")
plt.show()

# Correlation heatmap (optional)
plt.figure(figsize=(6,4))
sns.heatmap(df_sales.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
